# Querying no UniProt
Fazer buscas (Query) no banco de dados navegando na URL do site
Os parametros e formatos possíveis estão disponíveis nessa página:
[Retrieving entries via queries](https://www.uniprot.org/help/api_queries)


```
import requests

#Ponto de recuperação dos dados
urlBase = "https://www.uniprot.org/uniprot/"

#Parametros
payload = {
	"query": 'name:insulin AND reviewed:yes AND organism:"Homo sapiens (Human) [9606]"',
	"format": "list",
	"limit":"10",
	#"columns":"id,sequence",
	#"include":"yes",
	#"database":"pdb"
}

id=[]
result = requests.get(urlBase,params=payload)
if result.ok:
	print(result.url)
	print(result.text)
  #Transformando o resultado texto em lista
	id = result.text.split()
	print(id)
else:
	print("Algo errado!", result.status_code)
```

# Trabalhando com dados específicos
Você pode usar funções de manipulação de texto e laços de repetições para obter apenas os dados que te interessam.

```
#Exemplo: Querying Resultados
import requests

#Ponto de recuperação dos dados
urlBase = "https://www.uniprot.org/uniprot/"

#Parametros
payload = {
	"query": 'name:insulin AND reviewed:yes AND organism:"Homo sapiens (Human) [9606]"',
	"format": "txt",
	"limit":"1"
}

id=[]
result = requests.get(urlBase,params=payload)
if result.ok:
	print(result.text)
	for line in result.text.splitlines():
		if 'DR   GO;' in line:
			print(line)
else:
	print("Algo errado!", result.status_code)
```

# Mapeando Ids
Uma outra opssibilidade é você fazer o mapeamento de ID de identificação de proteínas de outros repositórios e o UniProt. Podendo ser:
Outros > UniProt
UniProt > Outros
Não é possível fazer o mapeamento/conversão de Outros > Outros

```
#Identificadores no formato necessário
ids= " ".join(id)
print(ids)


#Ponto de recuperação para Mapeamento de IDS (uploadlists)
urlBase = "https://www.uniprot.org/uploadlists/"

# Parametros necessários para a conversão de ids
payload = {
    "from": "ACC+ID",
    "to":"PDB_ID",
    "format":"list",
    "query": ids
}

result = requests.get(urlBase,params=payload)
if result.ok:

  #Transformando o resultado texto em lista
  resultado = result.text.split()
  print(resultado)
else:
	print("Algo errado!", result.status_code)

```

# Recuperando informações com SeqIO
A recuperação de informações pode ser feita de diferentes formas, uma delas é usando o SeqIO do BioPython. 
Dessa vez vamos utilizar uma outra biblioteca de acesso de URL para demonstrar outras possibilidades.
A Função SeqIO do Biopython tem capacidade de ler diferentes saídas de bancos de dados, inclusindo o formato XML do uniprot.

```
import requests

#Ponto de recuperação dos dados
urlBase = "https://www.uniprot.org/uniprot/"

#Parametros
payload = {
	"query": 'name:insulin AND reviewed:yes AND organism:"Homo sapiens (Human) [9606]"',
	"format": "xml",
	"limit":"1",
}

result = requests.get(urlBase,params=payload)
if result.ok:
	print(result.text)
else:
	print("Algo errado!", result.status_code)
```

```
# Abrindo arquivo XML
from Bio import SeqIO
import urllib

id = "Q9UQB8"
handle = urllib.request.urlopen("https://www.uniprot.org/uniprot/"+id+".xml")
record = SeqIO.read(handle, "uniprot-xml")
#print(record)
print(record.annotations['comment_subcellularlocation_location'])
```
